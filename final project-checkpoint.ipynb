# 2.3 Fazer as alterações necessárias

Value

```python
def troca(preco):
    multiplier = preco[-1]
    preco = preco[1:-1]
    
    if multiplier == "M":
        return float(preco) * 10**(6)
    elif multiplier == "K":
        return float(preco) * 10**(3)

values = []
for v in fifa["Value"]:
    values.append(troca(v))
    
fifa["Value"] = values
```

Wage

```python
wages = []
for w in fifa["Wage"]:
    wages.append(troca(w))

fifa["Wage"] = wages
```

Skills

```python
colunas = ['Unnamed: 0', 'Age', 'Overall', 'Potential', 'Value', 'Wage', 'Preferred Positions']

cols = [col for col in fifa.columns if col not in colunas]

for i in cols:
    fifa[i] = fifa[i].apply(lambda x: eval(x) if isinstance(x,str) else x)
```

Potential

```python
linhas = []
i = 0
while i < len(fifa["Unnamed: 0"]):
    linhas.append(fifa["Unnamed: 0"][i])
    i+=1
colunas = ['Overall', 'Potential']


overall_potential = fifa.loc[linhas, colunas]
    
    
potencial = []
for dado1 in overall_potential.Potential:
    potencial.append(dado1)

overall = []
for dado2 in overall_potential.Overall:
    overall.append(dado2)
    

evolution = []
i = 0
j = 0
while i < len(potencial):
    while j < len(overall):
        valor = potencial[i] - overall[j]
        evolution.append(valor)
        j+=1
        i+=1
    
evolucao = pd.Series(evolution)
fifa['Evolution'] = evolucao
```

```python
fifa = fifa.dropna()
```
